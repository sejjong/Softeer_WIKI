```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# CSV 파일 불러오기
df = pd.read_csv('mtcars.csv')

# 상위 5개 행 보기
df.head()
```

```python
# 조합별 개수
gear_trans = df.groupby(['gear', 'am']).size().reset_index(name='count')
print(gear_trans)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=gear_trans, x='gear', y='count', hue='am')
plt.title('# of Cars by Gear x Transmission')
plt.xlabel('Gear, Auto Transmission')
plt.ylabel('# of Cars')
plt.show()
```

```python
df.hist(figsize=(15, 10), bins=10)
plt.suptitle('Histograms of Variables', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(data=df, x='cyl', y='hp')
plt.title('Cylinder vs Horse Power')
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(data=df, x='mpg', y='hp')
plt.title('MPG vs Horse Power')
plt.show()
```

```python
corr_matrix = df.corr(numeric_only=True)
print(corr_matrix)

# (옵션) heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
